```python
class Bailarin:
    """
    Clase que representa a un bailarín con sus datos profesionales
    y sus acciones principales.
    """
    def __init__(self, nombre, estilo, años_experiencia):
        # 3 Atributos: Datos referentes al bailarín
        self.nombre = nombre
        self.estilo = estilo
        self.años_experiencia = años_experiencia

    # 3 Funciones: Acciones que realiza el bailarín
    def bailar(self):
        return f"¡{self.nombre} está ejecutando una coreografía de {self.estilo}!"

    def ensayar(self):
        return f"{self.nombre} está practicando en el estudio de danza."

    def estirar(self):
        return f"{self.nombre} está calentando sus músculos con {self.años_experiencia} años de técnica."

# Creación de los 2 objetos (instancias de la clase Bailarin)
bailarin1 = Bailarin("Carlos", "Salsa", 5)
bailarin2 = Bailarin("Ana", "Ballet", 10)

# Resultados y ejecución
objetos = [bailarin1, bailarin2]

for b in objetos:
    print(f"--- Perfil del Bailarín: {b.nombre} ---")
    print(f"Estilo: {b.estilo} | Experiencia: {b.años_experiencia} años")
    print(b.bailar())
    print(b.ensayar())
    print(b.estirar())
    print("-" * 45)

    print("Programa terminado")
    print("Programa realizado por Paola Silveyra NL = 50")
```
